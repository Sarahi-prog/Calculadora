#IMPLEMENTAR UNA RNA PARA MNIST UTILIZANDO LA PAUTA DEL PRIMER EJEMPLO.


```python
import numpy as np
import tensorflow as tf
from sklearn.metrics import accuracy_score, classification_report
# Importamos el MNIST clГЎsico de nГәmeros
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalizacion de datos
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)), # convierte la matriz en un vector plano
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

print(model.summary())
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
```

```python
# predic pide un lote de datos
y_pred = model.predict(x_test)
y_classes = np.argmax(y_pred, axis=1) # extrae la maxima probabilidad (de las 10 seleccionadas 1)

# Mostramos el reporte completo
print("Accuracy Final:", accuracy_score(y_test, y_classes))
print(classification_report(y_test, y_classes))
```
